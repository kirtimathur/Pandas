```python
import pandas as pd
df1 = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
#pd.merge(df1,df2) #df1 is left, df2 is right
#how: 'str' = 'inner', #which merging
    #on: 'IndexLabel | None' = None, ----which cols to be merged
```

```python
pd.merge(df1,df2,how="inner",on="key") #inner will give only common entries in both cols
#1st row diff for both cols
```

```python
pd.merge(df1,df2,how="left",on="key") #entire table of left(df1) and common in both
```

```python
pd.merge(df1,df2,how="right",on="key") #all keys of df2 and common
```

```python
pd.merge(df1,df2,how="outer",on="key") #all keys of left and right
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 
   
# Define a dictionary containing employee data 
data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
 
# Convert the dictionary into DataFrame  
df1 = pd.DataFrame(data1)
 
# Convert the dictionary into DataFrame  
df2 = pd.DataFrame(data2) 
```

```python
print(pd.merge(df1,df2,how="inner",on="key"),
      "\n\n",
pd.merge(df1,df2,how="outer",on="key"),
      "\n\n",
pd.merge(df1,df2,how="left",on="key"),
      "\n\n",
pd.merge(df1,df2,how="right",on="key"))
```
