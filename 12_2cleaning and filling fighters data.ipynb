```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv("C:\\Users\\asus\\OneDrive\\Desktop\\Python@ms\\data_sets\\upcoming-event.csv")
data.head()
```

```python
for i in data.columns:
    print(i,"-------",data[i].isnull().sum())
```

```python
#fixing criteria: del cols which have more than 70% missing values, using loop
for i in data.columns:
    if ((data[i].isnull().sum()*100)/data.shape[0])>=70:
        data.drop(columns=[i],inplace=True) #column
```

```python
for i in data.columns:
    print(i,"-------",data[i].isnull().sum())
```

```python
l=["b_dec_odds","b_sub_odds","b_ko_odds"]
for i in l:
    print(i,"----",data[i].dtypes)
```

```python
for i in data.columns:
    if ((data[i].isnull().sum()*100)/data.shape[0])>=1:
          data[i].fillna(data[i].median(),inplace=True)
```
