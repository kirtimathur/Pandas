```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'A': ['C0', 'C1', 'C2', 'C3'],
                          'B': ['D0', 'D1', 'D2', 'D3']})
```

```python
merge_left=pd.merge(df1,df2,how="left",on="key")
merge_left #made diff col names coz of ambiguity
```

```python
df11 = pd.DataFrame({'key': ['K0', 'K8', 'K2'],
                     'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']})
   
df22 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'c': ['C0', 'C1', 'C2', 'C3'],
                          'd': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(df11,df22,how="inner",on="key")
```

```python
pd.merge(df11,df22,how="outer",on="key")
```

```python
pd.merge(df11,df22,how="left",on="key")
```

```python
pd.merge(df11,df22,how="right",on="key")
```

```python
#merging on more than one keys: key1 and key2 as below
df1 = pd.DataFrame({'key1': ['K0', 'K8', 'K2', 'K3'],
                    'key2': ['K10', 'K18', 'K12', 'K13'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})
df2 = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                    'key2': ['K10', 'K11', 'K12', 'K13'],
                    'A': ['C0', 'C1', 'C2', 'C3'],
                    'B': ['D0', 'D1', 'D2', 'D3']})
```

```python
#k0 and k10 will be considered as single key (k8 and k18 not same in both so won't come in inner)
```

```python
pd.merge(df1,df2,how="inner",on=["key1","key2"])
```

```python
pd.merge(df1,df2,how="left",on=["key1","key2"])
```
