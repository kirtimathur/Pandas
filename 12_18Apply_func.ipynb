```python
#apply(): allows user to pass
import pandas as pd
data = pd.DataFrame({'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]})
data
```

```python
#task1:add the col name company length
import numpy as np
l=[]
for i in data["Company"]:
    l.append(len(i))
print(l)    

data["company_length"]=pd.Series(l)
data
```

```python
#alternative approach: instead of for loop use data["column name"]apply(functionality)
#extraction of new col from existing col is concept of feature engineering
```

```python
data["cl"]=data["Company"].apply(lambda x:len(x))
data
```

```python
#task2: to get a new col where names of all person comes in capital 
```

```python
data["upper"]=data["Person"].apply(lambda x:x.upper())
data
```

```python
#task3:to get log values of sales
import numpy as np
data["log"]=data["Sales"].apply(lambda x:np.log(x))
data
```

```python
#task4:to replace carl by charlie
data["Person"]=data["Person"].apply(lambda x:"charly" if x=="Carl" else x)
data
```
