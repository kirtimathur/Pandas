```python
import numpy as np
import pandas as pd
import seaborn as sns
df=pd.DataFrame({"A":[1,2,np.nan,5],"B":[5,np.nan,np.nan,7],"C":[1,2,np.nan,9]})
df
```

```python
df.dropna(axis=1,how="any") #axis=1 is col,deleting ones with 100% nan values, how="any" means if col has 
#atleast 1 missing values,deletes the entire col
#here, all cols have missing values
```

```python
df.dropna(axis=1,how="all") #all means if entire col has missing values, delete it 
df
```

```python
df.dropna(axis=0,how="all")
```

```python
employees=[("stuti",28,"varanasi"),
          ("saumya",32,"delhi"),
          ("aaditya",25,"mumbai"),
          ("saumya",32,"delhi"),
          ("saumya",32,"delhi"),
          ("saumya",32,"mumbai"),
          ("aaditya",25,"dehradun"),
          ("seema",32,"delhi")]
df=pd.DataFrame(employees,columns=["Name","Age","City"]) #dataframe created from list of tuple
df
```

```python
d=df.duplicated()  #df.duplicated--- keep: "Literal['first'] | Literal['last'] | Literal[False]" = 'first'
#1st entry treated as unique entry, after that duplicate
d
#masking, for entries
df[d]
```

```python
d=df.duplicated(keep="last") #4th entry is unique now
df[d]
```

```python
df=df.drop_duplicates() #by default, keep's value is 1st
df
```

```python
employees=[("stuti",28,"F",""),
          ("saumya",32,"F",""),
          ("aaditya",25,"","m"),
          ("saumya",32,"Fe",""),
          ("saumya",32,"Fe",""),
          ("saumya",32,"F",""),
          ("aaditya",26,"","m"),
          ("seema",32,"F","")]
df=pd.DataFrame(employees,columns=["Name","Age","isfemale","ismale"]) #dataframe created from list of tuple
df
```

```python
#STRUCTURAL ISSUE: make one col for both genders, feature engineering, making 1 col from existing cols
#1st check difference using unique(efficient in large data): to check inconsistency 
df["isfemale"].unique()
```

```python
df["isfemale"]=df["isfemale"].replace("Fe","F")
df["isfemale"]
```

```python
df["isfemale"]=df["isfemale"].replace("","m") #using equal to, to save changes
df["isfemale"]
```

```python
df.drop(columns=["ismale"],inplace=True)
df
```

```python
df.rename(columns={"isfemale":"gender"},inplace=True)
df
```
